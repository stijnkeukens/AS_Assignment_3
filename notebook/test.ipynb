```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
clean_df = pd.read_csv("../data/processed/network_AS3.csv")
```

```python
# Plot road, with different colors for links, bridges, and intersections
# Intersections are bigger
df_demo = clean_df
color_map = {'link': 'blue', 'bridge': 'orange', 'intersection': 'red', 'sourcesink': 'green'}
plt.figure(figsize=(12, 8))
for model_type, group in df_demo.groupby('model_type'):
    plt.scatter(group['lon'], group['lat'], 
                color=color_map[model_type], 
                label=model_type, 
                s=200 if model_type == 'intersection' else 100 if model_type == 'sourcesink' else 1, 
                alpha=0.7)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Road Network Visualization')
plt.legend()
```
